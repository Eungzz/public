```python
# 필요한 패키지 설치
!pip install nltk sklearn TextBlob

# 필요한 라이브러리 import
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
from sklearn.feature_extraction.text import TfidfVectorizer
from textblob import TextBlob
import pandas as pd
from google.colab import drive

nltk.download('punkt')
nltk.download('stopwords')

# Google Drive 마운트
drive.mount('/content/drive')

# 파일 로드
with open('/content/drive/My Drive/engtxt.txt', 'r') as file:
    data = file.read().replace('\n', '')

# 상위 10문장 출력 (문장 토큰화 후)
sentences = nltk.tokenize.sent_tokenize(data)
print(sentences[:10])

```

```python
# 텍스트 토큰화 및 불용어 처리
words = word_tokenize(data)
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.isalnum() and word not in stop_words]
```

```python
# 고빈도 단어 1500개 추출 및 빈도수 표시
word_freq = Counter(filtered_words)
common_words = word_freq.most_common(1500)
print(common_words)
```

```python
# 필요한 패키지 설치
!pip install pyphen

# 필요한 라이브러리 import
import pyphen
import re

# Pyphen 객체 생성 (음절 분리)
dic = pyphen.Pyphen(lang='en')

# 난이도 설정 사전
difficulty_scores = {
    'C': 0.1, 'CV': 0.1, 'VC': 0.1,
    'CVC': 0.2, 'VCV': 0.2,
    'CCV': 0.3, 'VCC': 0.3, 'CCVV': 0.3, 'VVCC': 0.3,
    'CCVC': 0.4, 'CVCC': 0.4, 'CCVCC': 0.4, 'CCVVC': 0.4, 'CVVCC': 0.4, 'CCVVCC': 0.4,
    'CCCV': 0.5, 'CCCVC': 0.5, 'CCCVV': 0.5, 'CCCVVC': 0.5,
    'CCCVCCC': 0.5, 'CCCVVCCC': 0.5
}


# 패턴 매치를 위한 함수
def pattern(word):
    pattern = ""
    for letter in word:
        if letter in 'aeiou':
            pattern += 'V'
        else:
            pattern += 'C'
    return pattern

# 난이도 계산 함수
def difficulty(word):
    patt = pattern(word)
    try:
        score = difficulty_scores[patt]
    except KeyError:
        score = 0.6
    if re.search('([aeiou]{2,})', word):
        score += 0.1
    return score

# 결과를 저장할 데이터프레임
df = pd.DataFrame(columns=['Word', 'Frequency', 'Syllable_Count', 'Pattern', 'Difficulty'])

for word, freq in common_words:
    df = df.append({
        'Word': word,
        'Frequency': freq,
        'Syllable_Count': len(dic.inserted(word).split('-')),
        'Pattern': pattern(word),
        'Difficulty': difficulty(word)
    }, ignore_index=True)

# 음절 수와 난이도에 따른 정렬
df = df.sort_values(['Syllable_Count', 'Difficulty'], ascending=[True, True])

# 결과 출력
print(df.head(10))
```

```python
def assign_difficulty(row):
    if row['Syllable_Count'] == 1 and row['Difficulty'] <= 0.2:
        return '아주 낮은 난이도'
    elif row['Syllable_Count'] in [1, 2] and row['Difficulty'] <= 0.4:
        return '낮은 난이도'
    elif row['Syllable_Count'] in [2, 3, 4] and row['Difficulty'] <= 0.4:
        return '중간 난이도'
    elif row['Syllable_Count'] in [2, 3] and row['Difficulty'] > 0.4:
        return '높은 난이도'
    elif row['Syllable_Count'] in [2, 3, 4] and row['Difficulty'] > 0.6:
        return '아주 높은 난이도'
    else:
        return '기타'

df['Difficulty_Group'] = df.apply(assign_difficulty, axis=1)

# 결과 출력
print(df.head(10))
```

```python
# '아주 낮은 난이도'와 '낮은 난이도'에 해당하는 단어만 추출
df_final = df[(df['Difficulty_Group'] == '아주 낮은 난이도') | (df['Difficulty_Group'] == '낮은 난이도')]

# 제거할 단어 리스트
remove_words = ['I', 'A', 'B', 'p', 'V', 'Ah', 'La']

# 주어진 단어 제거
df_final = df_final[~df_final['Word'].isin(remove_words)]

import nltk
import pandas as pd
import itertools
from nltk.corpus import cmudict

nltk.download('cmudict')

pron_dict = cmudict.dict()

# Get words from cmudict
words_from_cmudict = set(pron_dict.keys())

# Change to set for better performance
df_final_word_set = set(df_final['Word'])

def get_minimal_pairs(word):
    if word not in pron_dict:  # check if the word is in the dictionary
        return []
    word_pron = pron_dict[word][0]  # 가장 대표적인 발음을 사용
    minimal_pairs = []

    # Only compare with words in df_final
    for w in df_final_word_set:
        if w == word or w not in pron_dict:  # 같은 단어이거나 발음이 사전에 없는 경우 제외
            continue
        w_pron = pron_dict[w][0]  # 가장 대표적인 발음을 사용
        if len(w_pron) != len(word_pron):  # 발음 길이가 다른 경우 제외
            continue
        diff = sum(a != b for a, b in zip(word_pron, w_pron))  # 음소가 다른 개수 계산
        if diff == 1:  # 음소가 한 개만 다른 경우
            minimal_pairs.append(w)
        if len(minimal_pairs) >= 5:  # 최대 5개의 쌍만 저장
            break
    return minimal_pairs

# Add minimal pairs to dataframe
df_final['Minimal Pairs'] = df_final['Word'].apply(get_minimal_pairs)

# Save to Excel
df_final.to_excel('/content/drive/My Drive/minimal_pairs.xlsx', index=False)
```

```python
import openai

generated_sentences = set()  # 중복을 방지하기 위해 set 사용
attempt = 0  # 문장 생성 시도 횟수
simple_words_limit = 0.2  # 간단한 단어(예: be 동사, 전치사 등)의 비율 제한
sorted_df_words_limit = 0.8  # sorted_df에 포함된 단어의 비율 제한

simple_words_set = {"be", "do", "does", "am", "is", "are", "was", "were", "in", "on", "at", "by", "to", "under", "not"}
sorted_df_words_set = set(sorted_df['Word'])  # sorted_df['Word']는 sorted_df에서 단어를 나타내는 열입니다

while len(generated_sentences) < 300 and attempt < 1000:  # 최대 1000번 시도
    prompt = "Generate a short English sentence within 2 to 5 words."
    generated = openai.Completion.create(
        engine="text-davinci-003",
        prompt=prompt,
        max_tokens=10,  # 최대 토큰 수를 감소하여 단어 수 제한에 맞춤
        temperature=0.8,
        n=3,
        stop=None
    )

    for choice in generated.choices:
        sentence = choice.text.strip()
        word_count = len(sentence.split())
        if word_count < 2 or word_count > 5:  # 단어 수가 2와 5 사이가 아니면 건너뜁니다
            continue

        simple_words = sum(word in simple_words_set for word in sentence.lower().split())
        sorted_df_words = sum(word in sorted_df_words_set for word in sentence.lower().split())
        if simple_words / word_count > simple_words_limit or sorted_df_words / word_count > sorted_df_words_limit:
            continue  # 간단한 단어의 비율이 제한을 초과하거나 sorted_df 단어의 비율이 제한을 초과하는 경우 건너뜁니다

        generated_sentences.add(sentence)  # 생성된 문장을 set에 추가합니다.

    attempt += 1

print(list(generated_sentences))
```

```python
import nltk
nltk.download('cmudict')
from nltk.corpus import cmudict

d = cmudict.dict()

def syllable_count(word):
    try:
        return [len(list(y for y in x if y[-1].isdigit())) for x in d[word.lower()]]
    except KeyError:
        # 만약 단어가 cmudict에 없는 경우, 단어의 길이를 1로 계산
        return [1]
```

```python
import pandas as pd

# 문장의 길이, 단어의 총 음절 수를 계산하는 함수 정의
def sentence_info(sentence):
    words = sentence.split()
    length = len(words)
    syllables = len(sentence.replace(" ", ""))  # space를 제외한 길이를 총 음절 수로 대체
    return length, syllables

# 데이터프레임 초기화
df = pd.DataFrame(columns=['Sentence', 'Length', 'Syllables', 'TF-IDF'])

for sentence in generated_sentences:
    length, syllables = sentence_info(sentence)
    # TF-IDF 계산
    tfidf = engtxt_tfidf_dict.get(sentence.lower(), 0)
    df = df.append({'Sentence': sentence, 'Length': length, 'Syllables': syllables, 'TF-IDF': tfidf}, ignore_index=True)

# 'Length'를 기준으로 데이터프레임 정렬
df = df.sort_values(by=['Length'])

# 결과 출력
print(df)
```

```python
import re

# 음절별 difficulty 수치를 계산하는 함수 정의
def calculate_difficulty(word, difficulty_scores):
    word_difficulty = 0
    word = re.sub(r'[aeiou]', 'V', word)  # 모든 모음을 'V'로 치환
    word = re.sub(r'[^V]', 'C', word)     # 모음이 아닌 모든 문자(자음)를 'C'로 치환
    for pattern, score in difficulty_scores.items():
        word_difficulty += word.count(pattern) * score
    return word_difficulty

# 문장 읽기 난이도를 계산하는 함수 정의
def sentence_reading_difficulty(sentence, difficulty_scores):
    words = sentence.split()
    sentence_difficulty = sum(calculate_difficulty(word, difficulty_scores) for word in words)
    return sentence_difficulty

# difficulty 수치 정의
difficulty_scores = {
    'C': 0.1, 'CV': 0.1, 'VC': 0.1,
    'CVC': 0.2, 'VCV': 0.2,
    'CCV': 0.3, 'VCC': 0.3, 'CCVV': 0.3, 'VVCC': 0.3,
    'CCVC': 0.4, 'CVCC': 0.4, 'CCVCC': 0.4, 'CCVVC': 0.4, 'CVVCC': 0.4, 'CCVVCC': 0.4,
    'CCCV': 0.5, 'CCCVC': 0.5, 'CCCVCC': 0.5, 'CCCVV': 0.5, 'CCCVVC': 0.5, 'CCCVVCC': 0.5,
    'CCCVCCC': 0.5, 'CCCVVCCC': 0.5
}

for sentence in generated_sentences:
    length, syllables = sentence_info(sentence)
    # 단어 읽기 난이도와 문장 읽기 난이도 계산
    word_difficulty = calculate_difficulty(sentence, difficulty_scores)
    sentence_difficulty = sentence_reading_difficulty(sentence, difficulty_scores)
    # TF-IDF 계산
    tfidf = engtxt_tfidf_dict.get(sentence.lower(), 0)
    df = df.append({
        'Sentence': sentence,
        'Length': length,
        'Syllables': syllables,
        'Word Difficulty': word_difficulty,
        'Sentence Difficulty': sentence_difficulty,
        'TF-IDF': tfidf
    }, ignore_index=True)

# 'Length'를 기준으로 데이터프레임 정렬
df = df.sort_values(by=['Length'])

# 결과 출력
print(df)
```

```python
def categorize_difficulty(row):
    score = 0
    if row['Length'] <= 2:
        score += 1
    elif 2 < row['Length'] <= 3:
        score += 2
    else:
        score += 3

    if row['Syllables'] <= 2:
        score += 1
    elif 2 < row['Syllables'] <= 6:
        score += 2
    else:
        score += 3

    if row['Sentence Difficulty'] <= 1.5:
        score += 1
    elif 1.5 < row['Sentence Difficulty'] <= 3:
        score += 2
    else:
        score += 3

    if score <= 4:
        return '하'
    elif 4 < score <= 6:
        return '중'
    else:
        return '상'

# 난이도 카테고리 추가
df['Difficulty Category'] = df.apply(categorize_difficulty, axis=1)

# 카테고리별로 분리
df_low = df[df['Difficulty Category'] == '하']
df_medium = df[df['Difficulty Category'] == '중']
df_high = df[df['Difficulty Category'] == '상']

# Excel writer 객체 생성
writer = pd.ExcelWriter('/content/gdrive/My Drive/df_categorized.xlsx', engine='xlsxwriter')

# 각 데이터프레임을 별도의 시트에 쓰기
df_low.to_excel(writer, sheet_name='Low Difficulty', index=False)
df_medium.to_excel(writer, sheet_name='Medium Difficulty', index=False)
df_high.to_excel(writer, sheet_name='High Difficulty', index=False)

# Excel writer 객체 저장
writer.save()

print("Saved the categorized DataFrame to df_categorized.xlsx")
```
